```python
claim_total = pd.get_dummies(claim_total, columns=['incident_type', 'property_damage', 'collision_type', 'incident_severity', 'police_report_available', 'auto_make', 'auto_model'])
```

```python
linearmodel.fit(X_train, y_train)
```
